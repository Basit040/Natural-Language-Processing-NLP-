```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten,Embedding, Activation, Dropout 
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalMaxPooling1D
import numpy as np
from numpy import array
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv('twitter4000.csv')
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
# Converting tweets into list
text= df['twitts'].tolist()
text
```

```python
y=df['sentiment']
```

```python
# We are converting the text in to some numerical data
token= Tokenizer()
token.fit_on_texts(text)
```

```python
token
```

```python
# Number of unique words in text
vocab_size= len(token.word_index)+1
vocab_size
```

```python
# now we will use these index of each unique word
token.word_index
```

```python
# Here is an example for first words list
x= ['i to the a and']
```

```python
token.texts_to_sequences(x)
```

```python
# Similarly implement this on text data
encoded_text= token.texts_to_sequences(text)
```

```python
encoded_text
```

```python
print(encoded_text)
# Now we can use this numerical data of text in keras
```

```python
# we are limiting the text sixe 120 only
max_length=120
X= pad_sequences(encoded_text, maxlen=max_length, padding='post')
```

```python
print(X)
```

```python
X.shape
```

```python
# X is a sequnce of words
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2, stratify=y)
```

```python
vec_size= 300

model= Sequential()
model.add(Embedding(vocab_size, vec_size, input_length=max_length))

model.add(Conv1D(64,8, activation='relu'))
model.add(MaxPooling1D(2))
model.add(Dropout(0.2))
          
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
          
model.add(Dense(16, activation='relu'))
          
model.add(GlobalMaxPooling1D())
          
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
%%time
model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
def get_encoded(x):
    x=token.texts_to_sequences(x)
    x=pad_sequences(x,maxlen=max_length, padding='post')
    return x
```
